```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'RSNA_ROUND-Loss_curve.csv'
data = pd.read_csv(file_path)

# Remove leading spaces from column names
data.columns = data.columns.str.strip()

# Extract the data
server_round = data['Server Round']
loss_cnn = data['Loss(DP with CNN)']
loss_resnet50 = data['Loss(DP with RESNET50)']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(server_round, loss_cnn, label='Loss(DP with CNN)', marker='o')
plt.plot(server_round, loss_resnet50, label='Loss(DP with RESNET50)', marker='x')
plt.xlabel('Server Round')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.grid(True)
plt.show()
```
